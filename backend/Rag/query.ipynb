```python
import os
from llama_cpp import Llama

# 1. Define the absolute path using a raw string for Windows
MODEL_PATH = r"A:\Programming-Language-\Local-LLM-PRoject\Programing\Testing-code\llama.cpp\models\phi-3-mini.gguf"

# 2. Strict path verification before instantiation
if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError(f"Missing model at: {MODEL_PATH}\nCheck if the file is downloaded and the name matches exactly.")

# 3. Load LLM
llm = Llama(
    model_path=MODEL_PATH,
    n_threads=8,
    n_ctx=2048
)
```

```python
# Load vector DB
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
db = Chroma(persist_directory="rag/db", embedding_function=embedding)
```

```python
# Load LLM
llm = Llama(
    model_path=MODEL_PATH,
    n_threads=8,
    n_ctx=2048
)
```

```python
query = input("Ask: ")

# Retrieve relevant docs
docs = db.similarity_search(query, k=2)

context = "\n".join([doc.page_content for doc in docs])

# Combine with prompt
prompt = f"""
Use the following context to answer:

{context}

Question: {query}
"""

# Generate answer
response = llm(
    prompt,
    max_tokens=200
)

print("\nAI Answer:\n", response["choices"][0]["text"])
```
